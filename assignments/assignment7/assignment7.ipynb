(c) Clean a column

<font size = "3">

- Import the dataset "data_raw/results.csv" to a new object "results"
- Our goal is to replace any string values in the column "milliseconds" <br>
to missing values (NaNs) and covert to numeric
- To do so
    - "Extract list of non-numeric values"
    - "Replace certain values"
    - "Convert column to numeric"
    - "Display"
 


```python
# Write your own code here
ms_str=results.query("milliseconds.str.isnumeric()==False")
list_unique=pd.unique(ms_str["milliseconds"])
list_unique
list_new=[np.nan]

results["milliseconds"]=results["milliseconds"].replace(list_unique,list_new)
results["milliseconds"]=pd.to_numeric(results["milliseconds"])
results.dtypes
```

(d) Groupby + Aggregate

<font size = "3">

- Compute the mean and standard deviation of "laps", <br>
grouping by "constructorId". Store the new dataset <br>
as "df_aggCon_pos"
- Sort the aggregate dataset in decending <br>
order of mean laps using ".sort_values()"



```python
# Write your own code
df_aggCon_pos = (results.groupby("constructorId")
                      .agg(mean_laps = ('laps','mean'),
                           sd_laps =   ('laps','std')))
df_aggCon_pos_sorted=df_aggCon_pos.sort_values(by='mean_laps',ascending=False)
df_aggCon_pos_sorted

```

(e) Query + Groupby + Aggregate

<font size = "3">

- Use ".query()" to subset observations with <br>
'race_id >= 150'
- Compute the mean and standard deviation of "position", <br>
grouping by "constructorId". Store the new dataset <br>
as "df_subAggCon_pos"


```python
position_str=results.query("position.str.isnumeric()==False")
list_unique1=pd.unique(position_str["position"])
list_unique1
list_new1=[np.nan]
results["position"]=results["position"].replace(list_unique1,list_new1)
results["position"]=pd.to_numeric(results["position"])
```

```python
# Write your own code
df_subAggCon_pos=(results.query("raceId >= 150")
                        .groupby("constructorId")
                            .agg(mean_pos=("position","mean"),
                                 sd_pos=("position","std")))

df_subAggCon_pos
```
